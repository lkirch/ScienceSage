### Examining chunks.jsonl

```python
import json
import pandas as pd
```

```python
with open("../data/chunks/chunks.jsonl", "r", encoding="utf-8") as f:
    chunk_dicts = [json.loads(line) for line in f]

df = pd.DataFrame(chunk_dicts)
print(f"Loaded {len(df)} chunks.")
df.info()
```
