### Check Wikipedia for How Much Info Is Available on My Topics

https://pypi.org/project/Wikipedia-API/

```python
import wikipediaapi
import requests
from IPython.display import Image, display
from sciencesage.config import TOPICS
```

```python
wiki = wikipediaapi.Wikipedia(user_agent='sciencesage', language='en')
```

```python
for topic in TOPICS:
    page = wiki.page(topic)
    exists = page.exists()
    is_category = page.ns == wikipediaapi.Namespace.CATEGORY
    num_links = len(page.links)
    print(f"Topic: {topic}")
    print(f"  Exists: {exists}")
    print(f"  Is Category: {is_category}")
    print(f"  Number of links: {num_links}")
    print("-" * 40)
```

```python
science_cat = wiki.page("Category:Science")

# List subcategories of "Category:Science"
for title, page in science_cat.categorymembers.items():
    if page.ns == wikipediaapi.Namespace.CATEGORY:
        print(title)
```
